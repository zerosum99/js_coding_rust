## Scalar Types

- Floating point: f32, f64
- char Unicode scalar values like 'a', 'α' and '∞' (4 bytes each)

## 1-1 원시자료형 중에 스칼라 자료형인 정수: 부호로 자료형을 분리  

## 부호를 가지는 정수 알아보기

- i : 부호를 가지는 자료형
- u : 부호가 없는 자료형
- 그 다음에 비트 길이를 지정해서 변수를 정의 

```rust
i8::MIN
```

```rust
i8::MAX
```

### 출력되는 길이를 10으로 맞춤 

```rust
fn main() {
    println!(" {:10}, {:10} ",i8::MIN,i8::MAX); 
    println!(" {:10}, {:10} ",u8::MIN,u8::MAX); 

    println!("Success!");
}

main(); 
```

## 원시 자료형 내의 메서드 확인

-   원시자료형도 타입이라서 데이터 구조를 가지고
-   기본적으로 메서드들을 제공한다.

```rust
let x = 10i32;

println!("{}", x.abs());
println!("{}", x.count_ones());
println!("{}", x.leading_zeros());
println!("{}", x.pow(3));
println!("{}", x.to_string());
```

## 1-2 정수 자료형의 크기

- 자료형을 비트 사이즈에 맞춰 정의되어 있다.
- isize는 플랫폼에 맞춰 사이즈를 결정해서 64비트 os를 사용해서 8바이트로 처리 

### 크레이트 내의 모듈 사용하기

- use 키워드 :  사용 여부 결정
- std 는 공통 크레이트
- :: 는 접근
- mem 은 모듈

use std::mem;

### 모듈 내의 함수 사용

mem::size_of::<자료형>();

-  모듈명 :;  함수명 :;<자료형>()  

```rust
use std::mem;

fn main() {
    let size_of_u8 = mem::size_of::<u8>();
    let size_of_i32 = mem::size_of::<i32>();
    let size_of_isize = mem::size_of::<isize>();

    println!("u8 크기: {} 바이트", size_of_u8);
    println!("i32 크기: {} 바이트", size_of_i32);
    println!("isize 크기: {} 바이트", size_of_isize);
}

main();
```

## 1-3 엄격한 타입 체크 : 프리미티브 자료형도 동일한 자료형이 아니면  에러 처리 

- 엄격하게 자료형을 체크하므로 같은 정수여도 길이가 부호가 다르면 다른 자료형으로 인식
- 타입변환은 as 키워드 사용 함 

```rust
fn main() {
    let x: i32 = 5;
    let mut y: u32 = 5;

    y = x as u32;            // 명시적인 타입 변환 처리
    
    println!("Success!");
}
main();
```

### 자료형은 동일해야 함 

```rust
fn main() {
    let x: i32 = 5;
    let mut y: i32 = 5;

    y = x;
    
    println!("Success!");
}
main();
```

## 타입을 확인하는 함수 정의 

```rust
use std::any::type_name;

fn type_of<T>(_: T) -> &'static str {
    type_name::<T>()
}
```

## 정수에 대한 기본 타입 추론

- 변수에 자료형을 지정하지 않고 정수를 할당하면 기본 자료형은 i32 

```rust
fn main() {
    
    let z = 10;                       // Type of z ? 
    println!("타입: {}",type_of(z));
    
    let value = 42;
    let value_type = type_of(value);
    
    println!("값: {}", value);
    println!("타입: {}", value_type);

    println!("Success!");
}
main();
```

## 1-4 정수 오버플로우 확인하기 

- 정수 내에 있는 함수 호출 ( 정수자료형 :: 함수명(인자)
- result 이넘으로 결과를 반환하므로 unwrap 메서드로 결과값을 조회
- 결과가 에러이면 panic 을 발생시킴 

### 숫자에 접미사를 붙여서 자료형을 확정할 수 있다 

```rust
fn main() {
   //let v1 = 251_u8 + 8;
   let v2 = i8::checked_add(100, 8);
   let v3 = i8::checked_add(100, 8).unwrap();
   println!("{:?} {} ",v2, v3);
}

main();
```

### 오버플로우를 해결하기 위해서는 자료형을 변경 
 

```rust
fn main() {
   let v1 = 251_u16 + 8;
   let v2 = i16::checked_add(251, 8).unwrap();
   println!("{},{}",v1,v2);
}

main();
```

## 1-6 연산자 알아보기

-  기본적은 트레이트
-  이 트레이트를 자료형에 구현한 메서드가 연산자 임
- std::ops 모듈 내에 다양한 연산자를 가잠
- 

 ### 산술 연산자 트레이트
  트레이트	오버로드되는 연산자	시그니처
- Add, 	     a + b, 	              fn add(self, rhs) -> Output
- Sub,	a - b,	fn sub(self, rhs) -> Output
- Mul,	a * b,	fn mul(self, rhs) -> Output
- Div,	a / b,	fn div(self, rhs) -> Output
- Rem,	a % b,	fn rem(self, rhs) -> Output
- Neg,	-a,  	fn neg(self) -> Output
- Not,	!a,	    fn not(self) -> Output

```rust
// Fill the blanks and fix the errors
fn main() {
    
    let x = 1u32 + 2;   // 덧셈
    println!("덧셈 : {x}"); 
    let y = 1i32 - 2;   // 뺄셈
    println!("뺄셈 : {y}"); 
    let z = 3 * 50;     // 곱셈
    println!("곱셈 : {z}"); 
    let a = 9.6 / 3.2;  // 나눗셈 
    println!("나눗셈 : {a}"); 
    let b = 23 % 5;
    println!("나머지 : {b}"); 
} 

main(); 
```

## 비트 계산 및 출력 

- 이진수는 0b
- 8진수는 0o
- 16진수는 0x
- 출력포맷일 경우 : b(2진수), o(8진수), x(16진수)

### 비트 연산자
트레이트	연산자
- BitAnd	&
- BitOr	`
- BitXor	^
- Shl	    <<
- Shr	    >>

```rust
fn main() {    

    // Bitwise operations
    println!("0011 AND 0101 is {:04b}", 0b0011u32 & 0b0101);
    println!("0011 OR 0101 is {:04b}", 0b0011u32 | 0b0101);
    println!("0011 XOR 0101 is {:04b}", 0b0011u32 ^ 0b0101);
    println!("1 << 5 is {}", 1u32 << 5);
    println!("0x80 >> 2 is 0x{:x}", 0x80u32 >> 2);
}

main();
```

## 2-2 문자 

- 문자 자료형은 4 바이트로 처리 

### 값의 사이즈 게산 

```rust
use std::mem::size_of_val;
fn main() {
    let c1 = 'a';
    println!(" { }, {}", size_of_val(&c1), type_of(c1)); 

    let c2 = '中';
    println!(" { }, {}",size_of_val(&c2),type_of(c1)); 

    println!("Success!");
} 
main();
```
